```python
import numpy as np

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# weights
w1, w2, w3, w4 = 0.32, 0.41, 0.18, 0.27
w5, w6, w7, w8 = 0.12, 0.44, 0.35, 0.29

# bias
b1, b2 = 0.5, 0.7

# inputs
x1, x2 = 0.05, 0.1  

# outputs of hidden layers
h1_input = w1 * x1 + w2 * x2 + b1
h1 = tanh(h1_input)
print(f"h1 input: {h1_input:.4f} + h1 output: {h1:.4f}")

h2_input = w3 * x1 + w4 * x2 + b1
h2 = tanh(h2_input)
print(f"h2 input: {h2_input:.4f} + h2 output: {h2:.4f}")

# outputs of output layers
o1_input = w5 * h1 + w6 * h2 + b2
o1 = tanh(o1_input)
print(f"o1 input: {o1_input:.4f} + o1 output: {o1:.4f}")

o2_input = w7 * h1 + w8 * h2 + b2
o2 = tanh(o2_input)
print(f"o2 input: {o2_input:.4f} + o2 output: {o2:.4f}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

nodes = {
    "x1": (-1, 1), "x2": (-1, -1),  # المدخلات
    "h1": (0, 1), "h2": (0, -1),    # الطبقة المخفية
    "o1": (1, 0.5), "o2": (1, -0.5) # المخرجات
}

for node, pos in nodes.items():
    G.add_node(node, pos=pos)

edges = [
    ("x1", "h1", 0.32), ("x2", "h1", 0.41),
    ("x1", "h2", 0.18), ("x2", "h2", 0.27),
    ("h1", "o1", 0.12), ("h2", "o1", 0.44),
    ("h1", "o2", 0.35), ("h2", "o2", 0.29)
]

for src, dst, weight in edges:
    G.add_edge(src, dst, weight=weight)

pos = nx.get_node_attributes(G, 'pos')

plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", edge_color="gray", font_size=12, font_weight="bold")

edge_labels = {(src, dst): f"{weight:.2f}" for src, dst, weight in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, font_color="red")

plt.title("Neural Network Structure")
plt.show()
```
