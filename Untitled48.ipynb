```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)


X = [1, 2, 3]

input_dim = 1
hidden_dim = 1

Wf = 0.5
Wi = 0.6
Wo = 0.8
Wc = 0.7

Uf = 0.1
Ui = 0.2
Uo = 0.4
Uc = 0.3

bf = 0
bi = 0
bo = 0
bc = 0


h = 0
c = 0


print("------ LSTM Forward Pass ------\n")

for t, x in enumerate(X):

    print(f"Time Step {t+1}  |  Input = {x}")

    # Forget Gate
    f = sigmoid(Wf*x + Uf*h + bf)
    print("Forget Gate:", f)

    # Input Gate
    i = sigmoid(Wi*x + Ui*h + bi)
    print("Input Gate:", i)

    # Candidate Cell State
    c_tilde = tanh(Wc*x + Uc*h + bc)
    print("Candidate State:", c_tilde)

    # Cell State Update
    c = f*c + i*c_tilde
    print("Cell State:", c)

    # Output Gate
    o = sigmoid(Wo*x + Uo*h + bo)
    print("Output Gate:", o)

    # Hidden State
    h = o * tanh(c)
    print("Hidden State:", h)

    print("---------------------------")


Wy = 4.36
by = 0

y_pred = Wy*h + by

print("\nFinal Prediction:", y_pred)
```
